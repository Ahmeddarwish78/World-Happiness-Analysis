```python
cols = ['score'	,'gdp_per_capita',	'social_support',	'healthy_life_expectancy',	'freedom',	'generosity',	'corruption']

result = {}

for col in cols:
    highest = df.loc[df[col].idxmax(), 'country']
    lowest = df.loc[df[col].idxmin(), 'country']

    result[col] = {
        'highest': highest,
        'lowest': lowest
    }

result
summary = pd.DataFrame({
    'highest_country': [df.loc[df[col].idxmax(), 'country'] for col in cols],
    'lowest_country': [df.loc[df[col].idxmin(), 'country'] for col in cols]
}, index=cols)

summary
```

```python

lowest = df.loc[df['rank'].idxmin()]
highest = df.loc[df['rank'].idxmax()]
resulte=pd.concat([lowest, highest], axis=1)
resulte
```
